### Project Goal
This project builds an NLP system that automatically classifies short news articles into one of four topics: **World, Sports, Business,** or **Science/Technology**. The system follows the required NLP pipeline: problem definition, dataset preparation, text preprocessing, feature engineering, machine learning model training, evaluation, error analysis, and conclusion.


## 1. Problem Definition

### 1.1 What the Project Is Trying to Do
This project focuses on **news topic classification**, which is a supervised NLP task. The model learns from labeled news articles and predicts the correct topic category for new, unseen news text.

In simple terms, the project is trying to answer this question:

> Given a news title and description, can a machine learning model correctly identify whether the article is about World, Sports, Business, or Science/Technology?

### 1.2 Why the Problem Is Relevant
Online news platforms publish large amounts of articles every day. Manually sorting each article by topic takes time and may become inconsistent when different people label the same type of content differently. An automated classifier can help news websites, search engines, and recommendation systems organize articles faster and more consistently.

### 1.3 Input, Output, and Evaluation

| Component | Description |
|---|---|
| **Input data** | News article **Title + Description** combined into one text field |
| **Output prediction** | One of four topic labels: World, Sports, Business, or Sci/Tech |
| **NLP task type** | Multi-class text classification |
| **Main evaluation metric** | Accuracy |
| **Additional metrics** | Precision, Recall, F1-score, Confusion Matrix, and Error Analysis |

### 1.4 Class Labels

| Original Label | Topic Category |
|---:|---|
| 1 | World |
| 2 | Sports |
| 3 | Business |
| 4 | Sci/Tech |


## 2. Data Collection

### 2.1 Dataset Used
This project uses the **AG News Classification Dataset**, a public benchmark dataset commonly used for text classification experiments.

**Dataset source:** Kaggle, *AG News Classification Dataset*  
**Original reference:** Zhang, X., Zhao, J., & LeCun, Y. (2015). *Character-level convolutional networks for text classification.* Advances in Neural Information Processing Systems, 28.

### 2.2 Dataset Structure
The AG News dataset contains short news articles with three main columns:

| Column | Meaning |
|---|---|
| `Class Index` | Numeric topic label from 1 to 4 |
| `Title` | News headline/title |
| `Description` | Short news article description |

The standard AG News split contains:

| Split | Number of Samples | Class Balance |
|---|---:|---|
| Training set | 120,000 | 30,000 per class |
| Test set | 7,600 | 1,900 per class |

### 2.3 Why This Dataset Fits the Project
The dataset is appropriate because it is text-based, labeled, publicly available, and balanced across all four classes. It also goes far beyond the minimum sample requirement, making it suitable for supervised machine learning.

### 2.4 Data Preparation Plan
For this project, the title and description are combined into one text field because both contain useful information. The title often gives the main topic, while the description adds context that can help the model make a better prediction.


```python
# ============================================================
# CELL 1: Install and Import Libraries
# ============================================================
# If you are using Google Colab or a new environment, run this once:
# !pip install pandas numpy scikit-learn nltk matplotlib seaborn

import os
import re
import html
import warnings
from collections import Counter

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer, PorterStemmer

from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.metrics import (
    accuracy_score,
    classification_report,
    confusion_matrix,
    precision_recall_fscore_support
)
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC

warnings.filterwarnings("ignore")

# Download required NLTK resources
for package in ["punkt", "punkt_tab", "stopwords", "wordnet", "omw-1.4"]:
    try:
        nltk.download(package, quiet=True)
    except Exception:
        pass

print("Libraries imported successfully.")
```

```python
# ============================================================
# CELL 2: Load Dataset
# ============================================================
# Expected files:
# data/train.csv
# data/test.csv
#
# Kaggle AG News CSV files usually have no header and use columns:
# Class Index, Title, Description

TRAIN_PATH = "data/train.csv"
TEST_PATH = "data/test.csv"
COLUMNS = ["label", "title", "description"]

if not os.path.exists(TRAIN_PATH) or not os.path.exists(TEST_PATH):
    raise FileNotFoundError(
        "Please place AG News train.csv and test.csv inside a folder named 'data'. "
        "Expected paths: data/train.csv and data/test.csv"
    )

df_train = pd.read_csv(TRAIN_PATH, header=None, names=COLUMNS)
df_test = pd.read_csv(TEST_PATH, header=None, names=COLUMNS)

label_map = {1: "World", 2: "Sports", 3: "Business", 4: "Sci/Tech"}
df_train["category"] = df_train["label"].map(label_map)
df_test["category"] = df_test["label"].map(label_map)

print("Training set shape:", df_train.shape)
print("Test set shape:", df_test.shape)
df_train.head()
```

```python
# ============================================================
# CELL 3: Basic Data Cleaning Checks
# ============================================================

def dataset_report(df, name):
    print(f"--- {name} ---")
    print("Shape:", df.shape)
    print("Missing values:")
    print(df.isna().sum())
    print("Duplicate rows:", df.duplicated().sum())
    print("Class distribution:")
    print(df["category"].value_counts().sort_index())
    print()

dataset_report(df_train, "Training Data")
dataset_report(df_test, "Test Data")

# Drop missing and duplicate rows if any are found
before_train = len(df_train)
before_test = len(df_test)

df_train = df_train.dropna(subset=["label", "title", "description"]).drop_duplicates()
df_test = df_test.dropna(subset=["label", "title", "description"]).drop_duplicates()

print(f"Training rows removed: {before_train - len(df_train)}")
print(f"Test rows removed: {before_test - len(df_test)}")
```

```python
# ============================================================
# CELL 4: Visualize Class Distribution
# ============================================================

fig, axes = plt.subplots(1, 2, figsize=(13, 5))

for ax, df, title in zip(axes, [df_train, df_test], ["Training Set", "Test Set"]):
    counts = df["category"].value_counts().reindex(["World", "Sports", "Business", "Sci/Tech"])
    ax.bar(counts.index, counts.values)
    ax.set_title(f"Class Distribution: {title}")
    ax.set_xlabel("Category")
    ax.set_ylabel("Number of Articles")
    for i, value in enumerate(counts.values):
        ax.text(i, value, str(value), ha="center", va="bottom")

plt.tight_layout()
plt.show()
```

```python
# ============================================================
# CELL 5: Combine Title and Description
# ============================================================

def combine_text(row):
    return f"{row['title']} {row['description']}"

df_train["text"] = df_train.apply(combine_text, axis=1)
df_test["text"] = df_test.apply(combine_text, axis=1)

print("Example combined text:")
print(df_train["text"].iloc[0])
```

## 3. Text Preprocessing

### 3.1 Preprocessing Steps Used
Raw news text may contain uppercase/lowercase differences, punctuation, HTML entities, and very common words that do not strongly help topic prediction. To make the text cleaner and easier for machine learning models to process, the following preprocessing steps are applied:

| Step | What It Does | Why It Is Used |
|---|---|---|
| Lowercasing | Converts all text to lowercase | Prevents words like `News` and `news` from being treated as different words |
| HTML unescaping/removal | Removes web formatting and converts entities like `&amp;` | AG News articles come from online sources |
| URL removal | Removes website links | URLs usually add noise instead of useful topic meaning |
| Punctuation/special character removal | Keeps alphabetic words only | Helps BoW/TF-IDF focus on words |
| Tokenization | Splits sentences into tokens | Required before stopword removal and lemmatization |
| Stopword removal | Removes common words like `the`, `is`, `and` | Reduces low-information tokens |
| Lemmatization | Converts words to base form, such as `running` → `run` | Keeps words readable while reducing vocabulary size |

### 3.2 Why Lemmatization Was Chosen
Lemmatization was selected instead of stemming for the main pipeline because it keeps words more understandable. For example, stemming may reduce `businesses` into a shortened form, while lemmatization changes it into the valid word `business`. This makes the cleaned text easier to interpret during analysis.

### 3.3 Expected Impact
Preprocessing should reduce vocabulary size and remove noisy tokens. This can help models generalize better because the same idea is represented by fewer word variations.


```python
# ============================================================
# CELL 6: Define Text Preprocessing Function
# ============================================================

STOP_WORDS = set(stopwords.words("english"))
lemmatizer = WordNetLemmatizer()
stemmer = PorterStemmer()

def preprocess_text(text, method="lemmatize"):
    """
    Clean and normalize text for NLP classification.
    method: 'lemmatize', 'stem', or 'none'
    """
    text = str(text)
    text = html.unescape(text)                  # Convert HTML entities
    text = text.lower()                         # Lowercase
    text = re.sub(r"http\S+|www\.\S+", " ", text)  # Remove URLs
    text = re.sub(r"<.*?>", " ", text)         # Remove HTML tags
    text = re.sub(r"[^a-z\s]", " ", text)      # Keep letters and spaces only
    text = re.sub(r"\s+", " ", text).strip()  # Normalize spaces

    tokens = word_tokenize(text)
    tokens = [tok for tok in tokens if tok not in STOP_WORDS and len(tok) > 2]

    if method == "lemmatize":
        tokens = [lemmatizer.lemmatize(tok) for tok in tokens]
    elif method == "stem":
        tokens = [stemmer.stem(tok) for tok in tokens]

    return " ".join(tokens)

# Quick test
sample = "Breaking: Markets are rising after companies reported strong earnings!"
print("Original:", sample)
print("Cleaned :", preprocess_text(sample))
```

```python
# ============================================================
# CELL 7: Apply Preprocessing
# ============================================================
# Full AG News can take a few minutes. For faster testing, you may set SAMPLE_PER_CLASS.
# Use SAMPLE_PER_CLASS = None for full training data.

SAMPLE_PER_CLASS = None  # Example for fast testing: 7500

if SAMPLE_PER_CLASS is None:
    df_model_train = df_train.copy()
else:
    df_model_train = (
        df_train.groupby("label", group_keys=False)
        .apply(lambda x: x.sample(n=min(len(x), SAMPLE_PER_CLASS), random_state=42))
        .reset_index(drop=True)
    )

print("Training rows used:", len(df_model_train))
print("Test rows used:", len(df_test))

print("Preprocessing training data...")
df_model_train["text_clean"] = df_model_train["text"].apply(lambda x: preprocess_text(x, method="lemmatize"))

print("Preprocessing test data...")
df_test["text_clean"] = df_test["text"].apply(lambda x: preprocess_text(x, method="lemmatize"))

print("Preprocessing complete.")
df_model_train[["category", "text", "text_clean"]].head()
```

```python
# ============================================================
# CELL 8: Preprocessing Impact Analysis
# ============================================================

raw_tokens = []
clean_tokens = []

for text in df_model_train["text"]:
    raw_tokens.extend(word_tokenize(str(text).lower()))

for text in df_model_train["text_clean"]:
    clean_tokens.extend(str(text).split())

raw_vocab = len(set(raw_tokens))
clean_vocab = len(set(clean_tokens))
reduction = ((raw_vocab - clean_vocab) / raw_vocab) * 100 if raw_vocab else 0

impact_df = pd.DataFrame({
    "Metric": ["Total raw tokens", "Total clean tokens", "Raw vocabulary size", "Clean vocabulary size", "Vocabulary reduction (%)"],
    "Value": [len(raw_tokens), len(clean_tokens), raw_vocab, clean_vocab, round(reduction, 2)]
})

impact_df
```

```python
# ============================================================
# CELL 9: Most Common Clean Tokens by Category
# ============================================================

for category in ["World", "Sports", "Business", "Sci/Tech"]:
    tokens = []
    for text in df_model_train.loc[df_model_train["category"] == category, "text_clean"]:
        tokens.extend(text.split())
    print(f"
Top words for {category}:")
    print(Counter(tokens).most_common(15))
```

## 4. Feature Engineering

The cleaned text is converted into numerical features using **TF-IDF**. TF-IDF is used because it gives higher weight to words that are important in a document but not too common across all documents. This is useful for topic classification because words like `stocks`, `team`, `president`, or `software` can strongly indicate a specific category.

This notebook also compares a simple **Bag-of-Words + Naive Bayes** baseline with stronger TF-IDF-based models.


```python
# ============================================================
# CELL 10: Prepare Features and Labels
# ============================================================

X_train = df_model_train["text_clean"]
y_train = df_model_train["label"]

X_test = df_test["text_clean"]
y_test = df_test["label"]

print("X_train:", X_train.shape)
print("X_test :", X_test.shape)
```

## 5. Model Development

Three machine learning models are compared:

1. **Multinomial Naive Bayes** with Bag-of-Words as a simple baseline.
2. **Logistic Regression** with TF-IDF as a strong and interpretable classifier.
3. **Linear Support Vector Classifier** with TF-IDF, often effective for high-dimensional text data.

Basic hyperparameters:

| Model | Hyperparameter | Meaning |
|---|---|---|
| Naive Bayes | `alpha=1.0` | Smoothing value to handle unseen words |
| Logistic Regression | `C=2.0` | Controls regularization strength; higher values mean less regularization |
| Linear SVC | `C=1.0` | Controls margin/regularization balance |


```python
# ============================================================
# CELL 11: Train and Evaluate Models
# ============================================================

models = {
    "Bag-of-Words + Naive Bayes": Pipeline([
        ("vectorizer", CountVectorizer(max_features=50000, ngram_range=(1, 2))),
        ("classifier", MultinomialNB(alpha=1.0))
    ]),
    "TF-IDF + Logistic Regression": Pipeline([
        ("vectorizer", TfidfVectorizer(max_features=50000, ngram_range=(1, 2), sublinear_tf=True)),
        ("classifier", LogisticRegression(max_iter=1000, C=2.0, random_state=42))
    ]),
    "TF-IDF + Linear SVC": Pipeline([
        ("vectorizer", TfidfVectorizer(max_features=50000, ngram_range=(1, 2), sublinear_tf=True)),
        ("classifier", LinearSVC(C=1.0, random_state=42))
    ])
}

results = []
predictions = {}

for name, model in models.items():
    print(f"Training: {name}")
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    predictions[name] = y_pred

    precision, recall, f1, _ = precision_recall_fscore_support(
        y_test, y_pred, average="weighted", zero_division=0
    )
    acc = accuracy_score(y_test, y_pred)

    results.append({
        "Model": name,
        "Accuracy": acc,
        "Weighted Precision": precision,
        "Weighted Recall": recall,
        "Weighted F1-score": f1
    })

results_df = pd.DataFrame(results).sort_values(by="Weighted F1-score", ascending=False)
results_df
```

## 6. Evaluation and Analysis

The models are evaluated using accuracy, precision, recall, F1-score, confusion matrix, and misclassified examples. Accuracy gives the overall percentage of correct predictions, while precision, recall, and F1-score show how well the model performs across all classes.


```python
# ============================================================
# CELL 12: Detailed Evaluation of Best Model
# ============================================================

best_model_name = results_df.iloc[0]["Model"]
best_predictions = predictions[best_model_name]

target_names = [label_map[i] for i in sorted(label_map.keys())]

print("Best model:", best_model_name)
print("
Classification Report:")
print(classification_report(y_test, best_predictions, target_names=target_names))
```

```python
# ============================================================
# CELL 13: Confusion Matrix
# ============================================================

cm = confusion_matrix(y_test, best_predictions, labels=[1, 2, 3, 4])

plt.figure(figsize=(7, 5))
sns.heatmap(cm, annot=True, fmt="d", xticklabels=target_names, yticklabels=target_names)
plt.title(f"Confusion Matrix: {best_model_name}")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.show()
```

```python
# ============================================================
# CELL 14: Error Analysis
# ============================================================

error_df = df_test.copy()
error_df["predicted_label"] = best_predictions
error_df["predicted_category"] = error_df["predicted_label"].map(label_map)
error_df["correct"] = error_df["label"] == error_df["predicted_label"]

misclassified = error_df[~error_df["correct"]].copy()
print("Total misclassified examples:", len(misclassified))

misclassified[["category", "predicted_category", "title", "description"]].head(10)
```

### 6.1 Error Analysis Discussion

After viewing the misclassified examples above, common reasons for mistakes may include:

- Some articles contain overlapping vocabulary between **Business** and **Sci/Tech**, especially when technology companies, stocks, or product launches are discussed.
- Some **World** news articles mention sports teams, companies, or scientific topics, which can confuse the classifier.
- Short titles and descriptions may not provide enough context for the model.
- TF-IDF focuses on word frequency, so it may miss deeper meaning or context compared with transformer-based models.

These errors are expected in traditional machine learning text classification because the model mainly learns from word patterns rather than full sentence meaning.
